### **1. Convertir de Romano a entero**

```python
num='LVIII'

def romanToInt(s):

        o = {'M':1000,'D':500,'C':100,'L':50,'X':10,'V':5,'I':1}
        i = 1
        res=0
        for i in range(len(s)-1):
            if o[s[i]] < o[s[i+1]]:
                res-=o[s[i]]
            else:
                res+= o[s[i]]
        return res + o[s[-1]]

"""
Esta función convierte un número romano a un entero. 
Primero, se define un diccionario `o` que asocia cada símbolo romano con su valor numérico. 
Luego, se recorre la cadena `s` y se compara el valor del símbolo actual con el siguiente. 
Si el valor actual es menor que el siguiente, significa que debe restarse (por ejemplo, en "IV", 1 se resta de 5). 
De lo contrario, se suma al resultado. 
Al final, se agrega el valor del último símbolo porque no se compara con ningún otro. 
Es una forma eficiente de resolver este problema usando conceptos básicos de Python.
"""
print(romanToInt(num))
```

### **2. Convertir de Entero a romano**

```python
num=1994
def intToRoman( num):
        s={0:'M',1:'D',2:'C',3:'L',4:'X',5:'V',6:'I'}
        res,ini,l='',1000,0
        while l <=6:
            tmp = num // ini
            num-= ini*tmp
            if tmp<5:
                res = res + s[l]*tmp if tmp!=4 else res+ s[l] + s[l-1]
            else:
                if tmp==9:
                    res+= s[l] + s[l-2]
                else: 
                    res = res + s[l-1] + s[l]*(tmp-5) if tmp>5 else res + s[l-1]
            ini= ini // 10
            l+=2
        return res
"""
Esta función convierte un número entero a su representación en números romanos.
Primero, se define un diccionario `s` que asocia cada posición con su símbolo romano correspondiente.
Luego, se inicializan variables para el resultado (`res`), el valor inicial (`ini`) y un índice (`l`).
A continuación, se utiliza un bucle `while` para recorrer las posiciones de los números romanos.
En cada iteración, se calcula cuántas veces el valor actual cabe en el número (`tmp`).
Si `tmp` es menor que 5, se construye la representación romana correspondiente.
Si `tmp` es igual a 9, se añade el símbolo de la posición actual y el de dos posiciones atrás.
Si `tmp` es mayor que 5, se añade el símbolo de la posición anterior y se ajusta el número restante.
Finalmente, se actualizan las variables para la siguiente iteración.
El resultado final es la representación romana del número entero.
"""
print(intToRoman(num))
```

### **3. Prefijo común más largo**

```python
strs= ["reflower","flow","flight"]

def longestCommonPrefix(self, strs):
        menor = min(strs,key=len)
        if all(elem.startswith(menor) for elem in strs):
            return menor
        for j in range(len(menor)):
            pat=menor[:len(menor)-j]
            if pat!='' and all(elem.startswith(pat) for elem in strs):
                return pat
        return ''
"""
Esta función encuentra el prefijo común más largo entre una lista de cadenas.
Primero, se determina la cadena más corta de la lista `strs` utilizando `min` con el argumento `key=len`.
Luego, se verifica si todas las cadenas comienzan con esta cadena más corta utilizando `all` y `startswith`.
Si es así, se devuelve la cadena más corta como el prefijo común.
Si no, se itera sobre la longitud de la cadena más corta y se crea un patrón `pat` que es la cadena más corta menos un carácter al final.
Se verifica si `pat` no está vacío y si todas las cadenas comienzan con este patrón.
Si se encuentra un patrón que cumple con ambas condiciones, se devuelve como el prefijo común.
Si no se encuentra ningún prefijo común, se devuelve una cadena vacía.
"""
print(longestCommonPrefix(strs))
```

### **4. SUMA3**

```python
nums=[-1,0,1,2,-1,-4]
def threeSum(nums):
        nums.sort()
        res=[]
        n= len(nums)
        for i in range(n):
            if i >0 and nums[i]==nums[i-1]:
                continue
            a=nums[i]
            left,right=i+1,n-1
            while left < right:
                b,c = nums[left], nums[right]
                suma=a+b+c
                if suma==0:
                    res.append([a,b,c])
                    left+=1
                    right-=1
                    while left < right and nums[left]==nums[left-1]:
                        left+=1
                    while left < right and nums[right]==nums[right+1]:
                        right+=1
                elif suma < 0:
                    left+=1
                else:
                    right-=1
        return res
"""
Esta función encuentra todas las combinaciones únicas de tres números en una lista que suman cero.
Primero, se ordena la lista `nums` para facilitar la búsqueda de combinaciones.
Luego, se inicializa una lista `res` para almacenar los resultados y se obtiene la longitud de la lista `n`.
A continuación, se itera sobre cada número en la lista.
Si el número actual es igual al anterior, se omite para evitar duplicados.
Se establece un puntero izquierdo (`left`) y derecho (`right`) para buscar pares que sumen cero con el número actual.
Dentro de un bucle `while`, se calcula la suma de los tres números.
Si la suma es cero, se añade la combinación a `res` y se ajustan los punteros izquierdo y derecho.
Si la suma es menor que cero, se incrementa el puntero izquierdo para aumentar la suma.
Si la suma es mayor que cero, se decrementa el puntero derecho para disminuir la suma.
Finalmente, se devuelve la lista de combinaciones únicas que suman cero.
"""
print(threeSum(nums))
```

```python
nums=[-4,2,2,3,3,3]
objetivo=0

def threeSumClosest(nums, target):
        nums.sort()
        dist=2*10**4
        n=len(nums)
        for i in range(n):
            if i >0 and nums[i]==nums[i-1]:
                continue
            a=nums[i]
            left,right=i+1,n-1
            while left < right:
                
                t=a+nums[right]+nums[left]
                dif = target-(t)
                tpd = abs(dif)
                if dif==0:
                    return t
                elif dif>0:
                    if dist>tpd:
                        dist=tpd
                        suma=t
                    left+=1
                    while left < right and nums[left]==nums[left-1]:
                        left+=1
                else:
                    if dist>tpd:
                        dist=tpd
                        suma=t
                    right-=1
                    while left < right and nums[right]==nums[right+1]:
                        right-=1
        return suma
"""
Esta función encuentra la suma más cercana a un objetivo dado utilizando tres números de una lista.
Primero, se ordena la lista `nums` para facilitar la búsqueda de combinaciones.
Luego, se inicializa una variable `dist` para almacenar la distancia mínima encontrada y se obtiene la longitud de la lista `n`.
A continuación, se itera sobre cada número en la lista.
Si el número actual es igual al anterior, se omite para evitar duplicados.
Se establece un puntero izquierdo (`left`) y derecho (`right`) para buscar pares que se sumen al número actual.
Dentro de un bucle `while`, se calcula la suma de los tres números y la diferencia con el objetivo.
Si la suma es igual al objetivo, se devuelve la suma.
Si la suma es mayor que el objetivo, se ajusta el puntero derecho para disminuir la suma.
Si la suma es menor que el objetivo, se ajusta el puntero izquierdo para aumentar la suma.
En cada iteración, se verifica si la distancia a la suma actual es menor que la distancia mínima encontrada.
Si es así, se actualiza la distancia mínima y se guarda la suma actual.
Finalmente, se devuelve la suma más cercana encontrada.
"""
print(threeSumClosest(nums,objetivo))
```

### **6. SUMA4**

```python
nums=[1,0,-1,0,-2,2]
objetivo=0

def fourSum(nums, target):
        nums.sort()
        n=len(nums)
        res=[]
        for i in range(n):
            if i >0 and nums[i]==nums[i-1]:
                continue
            a=nums[i]
            for j in range(i+1,n):
                b=nums[j]
                left,right=j+1,n-1
                while left < right:
                    v = a+b+nums[left]+nums[right]
                    if v==target:
                        res.append([a,b,nums[left],nums[right]])
                        left+=1
                        right-=1
                        while left < right and nums[left]==nums[left-1]:
                            left+=1
                        while left < right and nums[right]==nums[right+1]:
                            right-=1
                    elif v<target:
                        left+=1
                    else:
                        right-=1
        return res
"""
Esta función encuentra todas las combinaciones únicas de cuatro números en una lista que suman un objetivo dado.
Primero, se ordena la lista `nums` para facilitar la búsqueda de combinaciones.
Luego, se inicializa una lista `res` para almacenar los resultados y se obtiene la longitud de la lista `n`.
A continuación, se itera sobre cada número en la lista como el primer número de la combinación.
Si el número actual es igual al anterior, se omite para evitar duplicados.
Luego, se itera sobre los siguientes números en la lista como el segundo número de la combinación.
Se establece un puntero izquierdo (`left`) y derecho (`right`) para buscar pares que sumen al objetivo
con los dos números actuales.
Dentro de un bucle `while`, se calcula la suma de los cuatro números.
Si la suma es igual al objetivo, se añade la combinación a `res` y se ajustan los punteros izquierdo y derecho.
Si la suma es menor que el objetivo, se incrementa el puntero izquierdo para aumentar la suma.
Si la suma es mayor que el objetivo, se decrementa el puntero derecho para disminuir la suma.
En cada iteración, se verifica si los punteros izquierdo y derecho apuntan a números duplicados y se ajustan en consecuencia.
Finalmente, se devuelve la lista de combinaciones únicas que suman al objetivo.
"""
print(fourSum(nums,objetivo))
```
